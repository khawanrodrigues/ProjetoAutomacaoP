<h2>Desafio:</h2>

Todos os dias, o nosso sistema atualiza as vendas do dia anterior. O seu trabalho diário, como analista, é enviar um e-mail para a diretoria, assim que começar a trabalhar, com o faturamento e a quantidade de produtos vendidos no dia anterior

E-mail da diretoria: seugmail+diretoria@gmail.com  
Local onde o sistema disponibiliza as vendas do dia anterior: https://drive.google.com/drive/folders/1AxkZklCpV-eAkcVWALBnHBlyFpMzDFqa

Para resolver isso, vamos usar o pyautogui, uma biblioteca de automação de comandos do mouse e do teclado

Comandos pyautogui: https://pyautogui.readthedocs.io/en/latest/quickstart.html

```python
# !pip install pyautogui
```

```python
# Primeiro passo: Entrar no Sistema

import pyautogui as py
import pyperclip as clip
import time

py.PAUSE = 1

py.hotkey("ctrl" , "t")
clip.copy("https://drive.google.com/drive/folders/1AxkZklCpV-eAkcVWALBnHBlyFpMzDFqa")
py.hotkey("ctrl" , "v")
py.press("enter")
time.sleep(4)


# Segundo passo: navegar até o local da base de dados

#py.click(x=178, y=265) # clicando no local
#time.sleep(4)

# Terceiro passo: fazer download da base de dados

py.click(x=94, y=260) # clique no download
time.sleep(4)
py.click(x=623, y=448) # clique no salvar
time.sleep(6)
```

<h3>Analisando indicados do banco de dados:</h3>

```python
# Quarto passso: fazer a analise dos indicativos
import pandas as pd

tabela= pd.read_excel(r"C:\Users\Khawan\Downloads\Vendas - Dez.xlsx")
display(tabela)

quantidade = tabela["Quantidade"].sum()
faturamento = tabela["Valor Final"].sum()
```

<h3>Enviando email pelo hotmail:</h3>

```python
# Quinto passo: enviar o email para a diretoria

# Abrindo o email
py.hotkey("ctrl" , "t")
py.write("https://outlook.live.com/mail/0/")
py.press("enter")
time.sleep(4)

# Escrevendo email
py.click(x=178, y=137)
py.write("khawan.rodrigues.santos@hotmail.com") # colocando o email
py.press("tab") # confirmando o email
py.press("tab")
clip.copy("Relatório Vendas") # escrevendo o assunto
py.hotkey("ctrl" ,"v")
py.press("tab")

#escrevendo o corpo do email
texto = f"""Prezados, Bom dia
O faturamento de ontem foi: R${faturamento:,.2f}
A quantidade de vendas de ontem foi: {quantidade:,}

Cordialmente,
Khawan
"""

py.write(texto)

py.hotkey("ctrl" , "enter") # apertando enter
```
